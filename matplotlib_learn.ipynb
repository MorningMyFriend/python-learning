```python
#coding=utf-8
import numpy as np
import matplotlib.pyplot as plt
```

```python

t = np.arange(0., 5., 0.2)

# figure和subplot 不写也会自动创建
# plt.figure()
# plt.subplot()

# 设定 x y 轴显示范围
plt.axis([0, 6, 0, 25])

# plt.plot(x, y, format, x2, y2, format2 ...)
# @format: 颜色+形状
# 颜色：r(red) b(blue) g(green) y(yellow) w(white) k(black) c(cyan)
# 形状：虚线-- 实线- dash dot line -. dotted line : 星* 三角> < ^ v 圈o 点. 方块s 

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^', t, t**0.5, 'co')
plt.show()
```

```python
# 散点图

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# plt.scatter(x, y, c, s, data=None)
# @c: color
# @s: point size
# @data: 字典，当给出data时，参数按照字典解析

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
# 画多张图
names = [s for s in 'abcdefghijklmn']

values = np.random.random(len('abcdefghijklmn'))*100

# plt.figure(num=None, figsize=None, dpi=None, facecolor=None, edgcolor=None)
# 画布大小调整： @figsize  @dpi
plt.figure(figsize=(19,11))

# plt.subplot(nrows, ncols, index)
# 当行列数小于10时， subplot(1,2,4) 等价于 subplot('124')

# 柱状图
plt.subplot(2,2,1)
plt.bar(names, values, width=0.9, align='center')

plt.subplot(2,2,2)
plt.scatter(names, values)
# 文字标注: 坐标(x,y)是网格上的值为参考
plt.text(6,80, '({0},{1:.1f})'.format(6, values[6]))


plt.subplot(2,2,3)
plt.plot(names, values, '-.b')
# 横纵坐标 label
plt.xlabel('characters')
plt.ylabel('y-values')
# 坐标轴范围
plt.xlim(-1,15)
plt.ylim(-10,100)
# 调整坐标轴刻度: {"linear", "log", "symlog", "logit", ...}
plt.yscale('symlog')
# 标题
plt.title('ysacle:symlog')
# 网格
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(names, values, '-.b')
# 横纵坐标 label
plt.xlabel('characters')
plt.ylabel('y-values')
# 坐标轴范围
plt.xlim(-1,15)
plt.ylim(-10,100)
# 调整坐标轴刻度: {"linear", "log", "symlog", "logit", ...}
plt.yscale('linear')
# 标题
plt.title('yscale:linear')
# 网格
plt.grid(True)



# 整体标题
plt.suptitle('suptitle')
plt.show()
```
